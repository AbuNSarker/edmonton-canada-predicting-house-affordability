# Structural type of dwelling and household size: Canada, provinces and territories, census metropolitan areas and census agglomerations with parts

```python
struct_dwel_df = pd.read_csv('../input/structural-type-of-dwelling-household-size-canada/98100040.csv')
struct_dwel_df.head()
```

```python
struct_dwel_df.shape
```

```python
struct_dwel_df.info()
```

```python
# Finding number of unique values in each column

struct_dwel_df.nunique(dropna=True).sort_values()
```

Total household size should be enough. We dont want granular data for 1 person, 2 person, etc. So we can drop all the columns for 1,2,3,4,and 5+ persons and its corresponding index which has all null values

```python
# struct_dwel_df.drop(struct_dwel_df.columns[[6,16]], axis=1, inplace=True)
struct_dwel_df.drop(struct_dwel_df.iloc[:, 6:16], inplace = True, axis = 1)
```

```python
struct_dwel_df.info()
```

```python
struct_dwel_df.drop(struct_dwel_df.columns[[6, 8, 10]], axis=1, inplace=True)
```

```python
struct_dwel_df.info()
```

```python
#renaming the columns for readability
struct_dwel_df.rename(columns = {'Structural type of dwelling (9)':'Structural type of dwelling', 
                                 'Household size (8):Total - Household size[1]':'Total - Household',
                                 'Household size (8):Number of persons in private households[7]':'Number of persons in private households', 
                                 'Household size (8):Average household size[8]' : 'Average household size'},
                                 inplace = True)
```

```python
struct_dwel_df.head()
```

**The above table has Structural type of dwelling and household size: Canada, provinces and territories, census metropolitan areas and census agglomerations with parts**

```python
struct_dwel_df.to_csv('Structural-dwelling-household-size-edited.csv',index=False)
```

# Population estimates, canada provinces, quarterly data (year 1946 - 2022)

Note: Postcensal estimates are based on the 2016 Census counts adjusted for census net undercoverage (CNU) (including adjustment for incompletely enumerated Indian reserves (IEIR)) and the components of demographic growth that occurred since that census. Intercensal estimates are produced using counts from two consecutive censuses adjusted for CNU (including IEIR) and postcensal estimates.

The population growth, which is used to calculate population estimates, is comprised of the natural growth (Table 17100059), international migration (Table 17100040) and interprovincial migration (Tables 17100020 and 17100045).
